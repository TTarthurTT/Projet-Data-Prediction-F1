```python
import pandas as pd
import sklearn as pd

# Charger les données
df = pd.read_csv('csv/constructors.csv')

# Vérifier les premières lignes
print(df.head())

# Supprimer les lignes avec des valeurs manquantes
df_cleaned = df.dropna()

# Convertir une colonne en datetime
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'])

# Renommer les colonnes pour enlever les espaces et les majuscules
df_cleaned.columns = df_cleaned.columns.str.strip().str.lower().str.replace(' ', '_')

# Supprimer les doublons
df_cleaned = df_cleaned.drop_duplicates()

# Normaliser une colonne
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_cleaned[['prix']] = scaler.fit_transform(df_cleaned[['prix']])

# Résultat final
print(df_cleaned.head())
```
